# (ii)Convert the values of Centigrade degrees into Fahrenheit degrees and vice versa

```python
import numpy as np
fvalues = [0, 12, 45.21, 34, 99.91, 32]
F = np.array(fvalues)
print("Values in Fahrenheit degrees:")
print(F)
print("Values in  Centigrade degrees:")
print(np.round((5*F/9 - 5*32/9),2))
```

# 2.(i)Arithmetic operations with numpy

```python
import numpy as np

arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])
print("Array 1:",arr1)
print("Array 2:",arr2)
addition = np.add(arr1, arr2)
mul=np.multiply(arr1,arr2)
sub=np.subtract(arr1,arr2)
arr = np.array([-1, -2, 1, 2, 3, -4])
ab=np.absolute(arr)
print("Addition : ",addition)
print("Multiplication : ",mul)
print("Subtraction : ",sub)
print("Array 3 : ",arr)
print("Absolute value : ",ab)
```

# Probabilty distribution using histograms

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generate some data for this
# demonstration.
data = np.random.normal(170, 10, 250)

# Fit a normal distribution to
# the data:
# mean and standard deviation
mu, std = norm.fit(data)

# Plot the histogram.
plt.hist(data, bins=25, density=True, alpha=0.6, color='b')

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

plt.plot(x, p, 'k', linewidth=2)
title = "Fit Values: {:.2f} and {:.2f}".format(mu, std)
plt.title(title)

plt.show()
```

# (ii)Contour plots

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 50, 2)
y = np.arange(0, 50, 3)     # Creating 2-D grid of features
[X, Y] = np.meshgrid(x,y)
fig,ax = plt.subplots(1, 1)
Z = np.cos(X / 2) + np.sin(Y / 4)    # plots contour lines 
ax.contourf(X, Y, Z)                  #use contour()
ax.set_title('Contour Plot')
ax.set_xlabel('x axis') 
ax.set_ylabel('y axis')
plt.show()
```

# 5.Univariate analysis

```python
import pandas as pd

#create DataFrame
df = pd.DataFrame({'points': [1, 1, 2, 3.5, 4, 4, 4, 5, 5, 6.5, 7, 7.4, 8, 13, 14.2],
'assists': [5, 7, 7, 9, 12, 9, 9, 4, 6, 8, 8, 9, 3, 2, 6],
'rebounds': [11, 8, 10, 6, 6, 5, 9, 12, 6, 6, 7, 8, 7, 9, 15]})
print(df.head())

#calculate mean of 'points'
print("\n")
print("Mean of points")
print(df['points'].mean())

#calculate median of 'points'
print("\n Median of points")
print(df['points'].median())

#calculate standard deviation of 'points' 
print("\n Standard deviation")
print(df['points'].std())

#create frequency table for 'points' 
print("\n Frequency distribution table")
print(df['points'].value_counts())


#boxplot
import matplotlib.pyplot as plt
print("\n Boxplot")
print(df.boxplot(column=['points'], grid=False, color='black'))

#histogram
print("\n Histogram")
print(df.hist(column='points', grid=False, edgecolor='black'))

#using seaborn
import seaborn as sns
print("\n Seaborn")
print(sns.kdeplot(df['points']))
```

# 7.Perform appending, slicing, addition and deletion of rows with a Pandas DataFrame.

```python
import pandas as pd
player_list = [['M.S.Dhoni', 36, 75, 5428000],
			['A.B.D Villers', 38, 74, 3428000],
			['V.Kholi', 31, 70, 8428000],
			['S.Smith', 34, 80, 4428000],
			['C.Gayle', 40, 100, 4528000],
			['J.Root', 33, 72, 7028000],
			['K.Peterson', 42, 85, 2528000]]
df = pd.DataFrame(player_list, columns=['Name', 'Age', 'Weight', 'Salary'])
print("Initial data set \n")
print(df)
df1 = df.iloc[0:4]
print("\n After slicing \n")
print(df1)
print("\n Appending row \n")
df=df.append({"Name":"Sachin","Age":45,"Weight":70,"Salary":2452562},ignore_index=True)
print(df)
print("\n Dropping rows \n")
print(df.drop(df.index[[1]]))
```

# (ii)Use index label to delete or drop rows from a Pandas DataFrame

```python
import pandas as pd 
details = {'Name' : ['Ankit', 'Aishwarya', 'Shaurya', 'Shivangi'],
	'Age' : [23, 21, 22, 21],
	'University' : ['BHU', 'JNU', 'DU', 'BHU'],}

df = pd.DataFrame(details, columns = ['Name', 'Age', 'University'],
				index = ['a', 'b', 'c', 'd'])
print("Initial dataframe")
print(df)
# return a new dataframe by dropping a
# row 'c' from dataframe
update_df = df.drop('c')
print("\n After deletion")
print(update_df)
#delete two rows
print("\n Deleting 2 rows")
update_df = df.drop(['b', 'c'])
print(update_df)
```

# 16. 3d plot

```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
fig = plt.figure(figsize = (5,5))
ax = plt.axes(projection='3d')
plt.show()
```

```python
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
t = np.arange(0, 10*np.pi, np.pi/50)
x = np.sin(t)
y = np.cos(t)

ax.plot3D(x, y, t)
ax.set_title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)

plt.show()
```
